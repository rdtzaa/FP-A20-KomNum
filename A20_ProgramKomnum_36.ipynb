```python

########################################
#                                      #
#  DEPARTEMEN TEKNIK INFORMATIKA       #
#  PROGRAM STUDI TEKNIK INFORMATIKA    #
#  Mata Kuliah [Komputasi Numerik - A] #
#  Final Project - Juni 2025           #
#                                      #
########################################

#-------------------------------------------------\
# GROUP: A-20                                     |
##################################################|
# 5025241070 - Azil Arliansyah Hidayat            |
# 5025241097 - Adriel Mahira Dharma               |
# 5025241119 - Naura Taqiyya Mazi                 |
# 5025241120 - Raditya Zhafran P.                 |
# 5025241121 - Riyannizaar Dwi A.                 |
#-------------------------------------------------/

import math

def factorial(n):
  return math.factorial(n)

def print_table(diff):
  print("Tabel Selisih (Δ):")
  for i in range(len(diff)):
    for j in range(len(diff[i])):
      print(f"diff[{j}, {i+1}] = {diff[i][j]:.2f}")
    print()

def NGB(x, y, x0, x_dicari):
  n = len(y)
  diff_table = [y[:]]

  for i in range(1,n):
    level = []
    prev = diff_table[-1]
    for j in range(len(prev)-1, 0, -1):
      val = prev[j] - prev[j - 1]
      level.insert(0, val)
    diff_table.append(level)

  print_table(diff_table[1:])

  h = x[1] - x[0]
  u = (x_dicari - x0) / h

  print(f"Step size h = {h}")
  print(f"u = (x - x0)/h = ({x_dicari} - {x0})/{h} = {u}\n")

  x0_index = x.index(x0)
  result = y[x0_index]
  print(f"f({x0}) = {result}")
  u_term = 1

  for i in range(1,x0_index+1):
    u_term *= (u+i-1)
    delta = diff_table[i][x0_index-i]
    temp = u_term / factorial(i)
    term = round(temp, 2) * delta
    print(f"Term for i={i}: {term:.2f} (u_term={u_term:.2f}, diff[{x0_index-i}, {i}]={delta})")
    result += term

  print(f"\nHasil Interpolasi Newton Gregory Backward untuk f({x_dicari}) = {result:.2f}")
  return result

# INPUTAN
x = list(map(int, input("Masukkan daftar x (pisahkan dengan spasi): ").split()))
y = list(map(int, input("Masukkan daftar y (pisahkan dengan spasi): ").split()))
x0 = int(input("Masukkan nilai x0: "))
x_dicari = int(input("Masukkan x yang ingin dicari: "))

NGB(x, y, x0, x_dicari)

# x = 2 4 6 8 10 12 14 16 18
# y = -940 -6008 -11652 1040 74020 279960 729932 1581088 3044340
# x0 = 10
# xt = 11



```
